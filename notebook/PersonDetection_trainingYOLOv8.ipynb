# Updating Ground Truth Labels for 'Person' in COCO Dataset

In this section, we'll demonstrate how to update the ground truth labels in a COCO dataset using a custom function. The goal is to focus on the 'person' label and modify the dataset accordingly.

## Load COCO Dataset

Before we proceed, make sure you have loaded your COCO dataset. If you haven't done so, refer to the previous sections on loading the training, validation, and testing datasets.

````python
def pick_person_label(dataset):
    """
    Update ground truth labels in the dataset, focusing only on 'person' class.

    Args:
    - dataset: FiftyOne dataset to be processed

    Returns:
    - None
    """
    # Iterate over the dataset
    for sample in dataset:
        # Get the detections
        detections = sample.ground_truth.detections

        # Filter out non-person detections
        detections = [detection for detection in detections if detection.label == "person"]

        # Update the detections
        sample.ground_truth.detections = detections

        # Save the sample
        sample.save()
````

````python
# Apply the label-picking function to the training dataset
pick_person_label(ds_train)

# Apply the label-picking function to the validation dataset
pick_person_label(ds_val)

# Apply the label-picking function to the testing dataset
pick_person_label(ds_test)
````

# Drawing Bounding Boxes on Images from Dataset using OpenCV

In this example, we'll demonstrate how to draw bounding boxes on images using OpenCV based on annotations in a dataset. We'll assume the dataset contains information about bounding box coordinates (e.g., xmin, ymin, xmax, ymax) for each object.

````python
# Select one random sample from the dataset
sample_to_display = ds_train.first()

# Access the bounding box of the first detection in the ground truth
bounding_box = sample_to_display.ground_truth.detections[0].bounding_box

# Print the bounding box
print("Bounding Box Coordinates:")
print(f"   - x_min: {bounding_box[0]}")
print(f"   - y_min: {bounding_box[1]}")
print(f"   - x_max: {bounding_box[2]}")
print(f"   - y_max: {bounding_box[3]}")
````

````python
import cv2
from google.colab.patches import cv2_imshow  # Required for displaying images in Colab

# Assuming you have already loaded your dataset 'ds_train'
# ...

# Select three random samples from the dataset
samples_to_display = ds_train.take(3)

# Iterate over the selected samples
for idx, sample in enumerate(samples_to_display):
    # Load the image
    image = cv2.imread(sample.filepath)

    # Get image dimensions
    height, width, _ = image.shape

    # Iterate over detections in the sample
    for detection in sample.ground_truth.detections:
        label = detection.label
        rel_box = detection.bounding_box

        # Convert relative coordinates to absolute coordinates
        x1 = int(rel_box[0] * width)
        y1 = int(rel_box[1] * height)
        x2 = int((rel_box[0] + rel_box[2]) * width)
        y2 = int((rel_box[1] + rel_box[3]) * height)

        # Draw bounding box on the image
        color = (0, 255, 0)  # Green color
        thickness = 2
        image = cv2.rectangle(image, (x1, y1), (x2, y2), color, thickness)
        cv2.putText(image, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, thickness)

    # Display the image
    print(f"Image {idx + 1}")
    cv2_imshow(image)
````

# Exporting Dataset to YOLO Format

To facilitate training with the YOLO (You Only Look Once) framework, you need to format your dataset in a way compatible with YOLO's requirements. Follow these steps to export your dataset to YOLO format:

## 1. Prepare Dataset

Ensure your dataset is loaded into FiftyOne, and each annotation includes bounding box information. This ensures that each object of interest in your dataset is appropriately labeled.

## 2. Export to YOLO Format

The YOLO format represents object annotations in a specific structure. Export your dataset to this format, where each annotation file corresponds to an image file and contains information about the labeled objects' positions and classes.

## 3. Organize Output

Create an output directory to store the YOLO format annotations. Each annotation file should be named after its corresponding image and contain the necessary information about object positions and classes.

## 4. Utilize YOLO for Training

With your dataset now in YOLO format, you are ready to proceed with training your object detection model using the YOLO framework. Follow the YOLO training procedures, considering your formatted dataset.

Adjust the steps based on your dataset specifics, and remember to adhere to the YOLO format conventions to ensure seamless integration with the training process.


````python
import fiftyone as fo
import fiftyone.utils.ultralytics as fou
import fiftyone.zoo as foz
````

````python
# The path to export the dataset
EXPORT_DIR = "/content/yolo_dataset"
````

````python
# YOLO format requires a common classes list
classes = ds_train.distinct("ground_truth.detections.label")

# Export Training Set
ds_train.export(
    export_dir=EXPORT_DIR,
    dataset_type=fo.types.YOLOv5Dataset,
    label_field="ground_truth",
    split="train",
    classes=classes,
)

# Export Validation Set
ds_val.export(
    export_dir=EXPORT_DIR,
    dataset_type=fo.types.YOLOv5Dataset,
    label_field="ground_truth",
    split="val",
    classes=classes,
)

# Export Test Set
ds_test.export(
    export_dir=EXPORT_DIR,
    dataset_type=fo.types.YOLOv5Dataset,
    label_field="ground_truth",
    split="test",
    classes=classes,
)
````

# YOLOv3 Metrics

The YOLOv3 object detection model produces several important metrics related to Mean Average Precision (mAP). Here are the key metrics:

- **`metrics_yolov8.box.map`**: Mean Average Precision over the range IoU (Intersection over Union) thresholds from 0.50 to 0.95 (mAP50-95).

- **`metrics_yolov8.box.map50`**: Mean Average Precision at IoU threshold 0.50 (mAP50).

- **`metrics_yolov8.box.map75`**: Mean Average Precision at IoU threshold 0.75 (mAP75).

- **`metrics_yolov8.box.maps`**: A list containing the Mean Average Precision values for each category over the range IoU thresholds from 0.50 to 0.95.

````python
print("mAP (50-95):", metrics_yolov8.box.map)
print("mAP50:", metrics_yolov8.box.map50)
print("mAP75:", metrics_yolov8.box.map75)
print("mAPs for each category:", metrics_yolov8.box.maps)
````
